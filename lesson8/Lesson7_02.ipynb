### Step 1 : Use list to get the json file

```python
import requests

url = 'https://data.moenv.gov.tw/api/v2/aqx_p_488?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=datacreationdate%20desc&format=JSON'

response = requests.get(url)
response.raise_for_status()
data = response.json()

records = []

for items in data['records']:
    records.append((
        items['sitename'],
        items['county'],
        items['aqi'],
        items['status'],                      
        items['pm2.5'],
        items['datacreationdate'],            
        items['latitude'],
        items['longitude']
    ))

print(records)
print(items)
```

### Step 2 : Connect to SQLite database (or create it) and create an empty table by Drop existing table

```python
import sqlite3

# Connect to SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect("AQI_EX.db")

# Create a cursor object
cursor = conn.cursor()

# Drop the table if it already exists for re-run purposes and avoid the duplicated data
cursor.execute("DROP TABLE IF EXISTS records")  #   ---------------------important code-----------------------
```

### Step 3 : Create the 'records' table and add an id field

```python
sql = '''
CREATE TABLE IF NOT EXISTS records (
	id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,   
	sitename TEXT NOT NULL,
	county TEXT,
	aqi INTEGER,
	status TEXT,
	pm25 NUMERIC,
	date TEXT,
	lat NUMERIC,
	lon NUMERIC 
);
'''
# Create a table
cursor.execute(sql)
```

### Step 4 : Insert the 'records' list data into the SQLite table

```python
# Insert records table by naming all field names because created table has 9 fields but records only has 8 fields
cursor.executemany("INSERT INTO records (sitename, county, aqi, status, pm25, date, lat, lon) VALUES (?, ?, ?, ?, ?, ?, ?, ?)", records)  #  ------------------important code-------------------------------


conn.commit()
conn.close()

print("Data inserted successfully into the 'records' table.")
```
