```python
import sqlite3
conn = sqlite3.connect("AQI.db")
sql = '''
CREATE TABLE IF NOT EXISTS records (
	id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
	sitename TEXT NOT NULL,
	county TEXT,
	aqi INTEGER,
	status TEXT,
	pm25 NUMERIC,
	date TEXT,
	lat NUMERIC,
	lon NUMERIC,
    UNIQUE(sitename,date)
);
'''
cursor = conn.cursor()
cursor.execute(sql)
conn.commit()
cursor.close()
conn.close()
```
